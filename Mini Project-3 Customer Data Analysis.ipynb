### Performing feature Engineering 

```python
# Create a new column 'Age' based on 'Year_Birth'
current_year = 2023
data['Age'] = current_year - data['Year_Birth']
data['Age']
```

```python
# Create a new column 'Total_Spent' based on the sum of spending on different products
product_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
data['Total_Spent'] = data[product_columns].sum(axis=1)
data['Total_Spent']
```

### Clustering the Data 

```python
# Select relevant columns for clustering
columns_for_clustering = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
                          'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
                          'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
```

```python
# Create a new dataframe with only the selected columns
data_for_clustering = data[columns_for_clustering].copy()
data_for_clustering
```

```python
# Scale the data to ensure all features have equal importance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_for_clustering_scaled = scaler.fit_transform(data_for_clustering)
data_for_clustering_scaled
```

### Determine The Optimal Number of Systems 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Initialize an empty list to store the silhouette scores
silhouette_scores = []
```

```python
# Try different values of k (number of clusters)
for k in range(2,11):
    
    
# Create a KMeans instance with k clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
```

```python
# Fit the model to the scaled data
kmeans.fit(data_for_clustering_scaled)
```

```python
# Calculate the silhouette score for the clustering
silhouette = silhouette_score(data_for_clustering_scaled, kmeans.labels_)
```

```python
# Append the silhouette score to the list
silhouette_scores.append(silhouette)
```

```python
# Plot the silhouette scores for different values of k
#plt.plot(range(2,11), silhouette_scores)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Values of k')
plt.show()
```

### Perform the Clustering with the choosen number of clusters 

```python
# Choose the optimal number of clusters based on the silhouette score plot
k = 3  # Example: Assume 3 clusters
```

```python
# Create a KMeans instance with the chosen number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans
```

```python
# Fit the model to the scaled data
kmeans.fit(data_for_clustering_scaled)
```

```python
# Assign cluster labels to the original dataset
data['Cluster'] = kmeans.labels_
data['Cluster']
```

### Analyze The Cluster

```python
# Analyze the characteristics of each cluster
cluster_characteristics = data.groupby('Cluster')[columns_for_clustering].mean()
```

```python
# Visualize the characteristics of each cluster
for column in columns_for_clustering:
    plt.figure(figsize=(5, 4))
    sns.barplot(x='Cluster', y=column, data=data)
    plt.xlabel('Cluster')
    plt.ylabel(column)
    plt.title(f'Mean {column} for Each Cluster')
    plt.show()
```

Here demonstrate a basic approach to clustering the customer data using K-means clustering and visualizing the characteristics of each cluster.

### Hierarchy clustering ( Hierarchical clustering )

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Select relevant columns for clustering
columns_for_clustering = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
                          'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
                          'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
```

```python
# Create a new dataframe with only the selected columns
data_for_clustering = data[columns_for_clustering].copy()
data_for_clustering
```

```python
# Standardize the data
scaler = StandardScaler()
data_for_clustering_scaled = scaler.fit_transform(data_for_clustering)
data_for_clustering_scaled
```

### Perform the Hierarchical clustering 

```python
# Compute the linkage matrix
Z = linkage(data_for_clustering_scaled, method='ward')
```

```python
# Plot the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python

import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
 
# randomly chosen dataset
X = np.array([[1, 2], [1, 4], [1, 0],
              [4, 2], [4, 4], [4, 0]])
 
# Perform hierarchical clustering
Z = linkage(X, 'ward')
 
# Plot dendrogram
dendrogram(Z)
 
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data point')
plt.ylabel('Distance')
plt.show()
```

### Determine the Optimal number of features 

```python
# Determine the optimal number of clusters using the elbow method
distances = Z[-10:, 2]
distances_rev = distances[::-1]
plt.plot(range(2, 12), distances_rev)
plt.xlabel('Number of Clusters')
plt.ylabel('Distance')
plt.title('Elbow Method')
plt.show()
```

```python
# Choose the optimal number of clusters based on the elbow method
k = 4  # Example: Assume 4 clusters
```

```python
# Perform clustering with the chosen number of clusters
from sklearn.cluster import AgglomerativeClustering
cluster_model = AgglomerativeClustering(n_clusters=k)
cluster_labels = cluster_model.fit_predict(data_for_clustering_scaled)
```

```python
# Assign cluster labels to the original dataset
data['Cluster'] = cluster_labels
```

```python
# Analyze the characteristics of each cluster
cluster_characteristics = data.groupby('Cluster')[columns_for_clustering].mean()
```

```python
# Visualize the characteristics of each cluster
for column in columns_for_clustering:
    plt.figure(figsize=(5,4))
    sns.barplot(x='Cluster', y=column, data=data)
    plt.xlabel('Cluster')
    plt.ylabel(column)
    plt.title(f'Mean {column} for Each Cluster')
    plt.show()
```

By performing hierarchical clustering and analyzing the characteristics of each cluster, you can gain insights into different customer segments based on their spending habits, recency of purchases, income level, and other attributes
